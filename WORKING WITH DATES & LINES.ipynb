WORKING WITH DATES CHANGING THEM FROM OBJECT

```python
#CONVERTING them into date time datatype
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

EXTRACTING YEAR FROM THIS DATE


```python
date['date_year']= date['date'].dt.year
date.head()
```

EXTRACTING MONTH

```python
date['date_month_no'] = date['date'].dt.month
date.head()
```

```python
date['date_month_name']= date['date'].dt.month_name()
date.head()
```

EXTRACTING EXACT DATE

```python
date['date_day']= date['date'].dt.day
date.head()
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

DAY OF WEEK

```python
date['date_day_name']= date['date'].dt.dayofweek
date.head()
```

WEEKEND

```python
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

WEEK OF YEAR

```python
date['date_week'] = date['date'].dt.isocalendar().week

date.drop(columns=['product_id','city_id','orders']).head()
```

QUATAR

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

EXTRACTING SEMESTER

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

MONTHS PASSED

```python
from dateutil.relativedelta import relativedelta

# Get today's date
today = datetime.datetime.today()

# Example: Replace this with the actual date you want to compare
given_date = datetime.datetime(2023, 1, 1)

# Calculate the number of months passed
months_passed = relativedelta(today, given_date).months + 12 * (today.year - given_date.year)
months_passed
```
